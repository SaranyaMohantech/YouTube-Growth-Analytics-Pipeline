```python
import pandas as pd
```

#Load the raw YouTube analytics export

```python
df = pd.read_csv('data\\raw\\youtube\\video_metrics.csv')
df = df[df['Content'] != 'Total'] #Drops the Total row from the data  
```

```python
print(df.info())
```

#Data cleaning: Strip trailing spaces from video titles and convert dates

```python
df['Video title']= df['Video title'].str.strip() #Removes leading and trailing spaces from the video title column
df['Video publish time'] = pd.to_datetime(df['Video publish time']) #Converts the video publish time column to datetime format
```

#Extract day of week and hour to locate the 'Golden Window"

```python
df['publish_day'] = df['Video publish time'].dt.day_name()
df['publish_hour'] = df['Video publish time'].dt.hour
```

Filter out unnecessary column and structure it exactly for snowflake table

#We map: Content -> video_id, video title -> video_title, watch time (hours) -> watch_time_hours, subscribers -> subscribers_gained, Publish_Day -> publish_day, publish_Hour -> publish_hour

```python
snowflake_df =df[[
    'Content', 
    'Video title', 
    'Views', 
    'Watch time (hours)', 
    'Subscribers', 
    'publish_day', 
    'publish_hour']].copy()
```

#Rename columns to match your lowercase snowflake schema exactly

```python
snowflake_df.columns = [
    'video_id', 
    'video_title', 
    'views', 
    'watch_time_hours', 
    'subscribers_gained', 
    'publish_day', 
    'publish_hour'
]
```

#Save the perfectly structured data

```python
snowflake_df.to_csv('data/raw/youtube/video_metrics_clean_snowflake.csv', index=False)
print("Pipeline complete: All errors resolved! File is ready for Snowflake.")
```
